2. We can use Normalization or Standardization to scale our data. Normalization (Min-Max scaling) typically rescales the values into a range of [0,1]. On the other hand, Standardization typically rescales data in such a way that the features will have the properties of a standard normal distribution (i.e. a mean of 0 and a standard deviation of 1).

    Normalization or Standardization? The answer of this question heavily relies on the application of our program.
    
    * Standardizing is important when we compare measurements that have different units. Variables that are measured at different scales do not contribute equally to the analysis and might end up creating a bais. Standardization assumes that your data has a Gaussian (bell curve) distribution. This does not strictly have to be true, but the technique is more effective if your attribute distribution is Gaussian. Standardization is useful when your data has varying scales and the algorithm you are using does make assumptions about your data having a Gaussian distribution, such as linear regression, logistic regression, and linear discriminant analysis.
    
    * Normalization is a good technique to use when you do not know the distribution of your data or when you know the distribution is not Gaussian (a bell curve). Normalization is useful when your data has varying scales and the algorithm you are using does not make assumptions about the distribution of your data, such as k-nearest neighbors and artificial neural networks.
    
    We can conclude that normalization is the better method for our usage, because we don't know the distribution. Also, we can handle our (release_year) range in a cleaner fashion this way.
    
    Further reading: $\href{https://towardsai.net/p/data-science/how-when-and-why-should-you-normalize-standardize-rescale-your-data-3f083def38ff}{link}$ and $\href{https://sebastianraschka.com/Articles/2014_about_feature_scaling.html#about-min-max-scaling}{link}$

3. To handle categorical data we can use the following methods:
    * Label Encoding: It consists of substituting each group with a corresponding number and keeping such numbering consistent throughout the feature. 
    * One-Hot Encoding: It's the most common, correct way to deal with non-ordinal categorical data. It consists of creating an additional feature for each group of the categorical feature and mark each observation belonging (Value=1) or not (Value=0) to that group.
    * Target Encoding: It consists of substituting each group in a categorical feature with the average response in the target variable.

    Usually, label encoding is not the proper solution, because numbers, themselves, have relations with eachother, but here since we don't rely on any relationship between the genres, We use label encoding to handle our data. Note that label encoding is done mainly in Part Two.
    Here, for a better demonstration, we have encoded the countries as shown below:
    
    Further reading: $\href{https://towardsdatascience.com/handling-categorical-data-the-right-way-9d1279956fc6}{link}$

4. As it is shown below, for each cell under the "listed_in" column, we have a specific list. We can either merge that list into a unique string, or we can parse the list and make new features. We use the latter to avoid overfitting. 

```python
dataframe["listed_in"]
```

```python
max_list_count = 0
def calc_genre_no_max(row):
    global max_list_count
    cell = row['listed_in']
    max_list_count = max(max_list_count, len(cell.split(",")))

dataframe.apply(calc_genre_no_max, axis=1)
print("Maximum number of items in each list is:", max_list_count)

def remove_comma(row):
    word = row['cast']
    return word.replace(" ", "")
    #return word.replace(",", " ")
dataframe['cast'] = dataframe.apply(remove_comma, axis=1) #remove comma from cast to generate a unique string for each name!
dataframe
```

So it seems like preserving 5 categories would be appropriate to handle the data, without generating alot of bias.

```python
for length in range(0, 5):
    new_genre = []
    for item in dataframe["listed_in"]:
        if len(item.split(",")) > length:
            new_genre.append(item.split(",")[length])
        else:
            index = length % len(item.split(","))
            new_genre.append(item.split(",")[index])
    column_name = "genre no." + str(length + 1)
    dataframe[column_name] = new_genre
    
```

```python
del dataframe["listed_in"]
dataframe
```

We repeat the same action with the cast column.

```python
avg_cast_count = 0
avg_cast_sum = 0
def calc_cast_no_avg(row):
    global avg_cast_count, avg_cast_sum
    cell = row['cast']
    avg_cast_sum += len(cell.split(","))
    avg_cast_count += 1

dataframe.apply(calc_cast_no_avg, axis=1)
print("Avg number of people in each cast is:", avg_cast_sum/avg_cast_count)
```

So we consider two positions for cast people as well.

```python
for length in range(0, 2):
    new_cast = []
    for item in dataframe["cast"]:
        if len(item.split(",")) > length:
            new_cast.append(item.split(",")[length])
        else:
            index = length % len(item.split(","))
            new_cast.append(item.split(",")[index])
    column_name = "cast no." + str(length + 1)
    dataframe[column_name] = new_cast
```

As it is shown, we have reached 98% accuracy using the following hyperparameters:

$\color{blue}{max-depth:}$ The maximum depth of the tree. If None, then nodes are expanded until all leaves are pure or until all leaves contain less than min_samples_split samples.

$\color{blue}{min-samples-leaf:}$ The minimum number of samples required to split an internal node.

$\color{blue}{min-samples-split:}$ The minimum number of samples required to be at a leaf node. A split point at any depth will only be considered if it leaves at least |min_samples_leaf| training samples in each of the left and right branches.

As it is shown, if we reduce the test size and spend more resources on training, overfitting would occur and our method would lose generality, so it wouldn't function well on a new data.
Similarly, if we increase the test size, accuracy will drop after a certain point (high bias).
In conclusion, we need to maintain a proper ratio of test to train data.

<h2> $Part \hspace{3pt} \color{green}{Three}:$ </h2>

A simple random forest (without any hyperparameter tuning) would look like below:

```python
from sklearn.ensemble import RandomForestClassifier

random_forest = RandomForestClassifier()
random_forest.fit(train_x, train_y)
predict_forest = random_forest.predict(test_x)
print("Accuracy is: ", accuracy_score(test_y, predict_forest))
    
  
```

Which is very accurate. We can examine the effect hyperparameters have on our accuracy:
1. n_estimators: number of trees in the foreset
2. max_features: max number of features considered for splitting a node
3. max_depth: max number of levels in each decision tree
4. min_samples_split: min number of data points placed in a node before the node is split
5. min_samples_leaf: min number of data points allowed in a leaf node

<h4> The effect of n_estimators on accuracy: </h4>

```python
n_estimators_vals = []
indices = []
for depth in range(95, 105):
    clf_model_n_estimators = RandomForestClassifier(n_estimators=depth)   
    clf_model_n_estimators.fit(train_x, train_y)
    predict_y_n_estimators = clf_model_n_estimators.predict(test_x)
    n_estimators_vals.append(accuracy_score(test_y, predict_y_n_estimators))
    indices.append(depth)
plt.plot(indices, n_estimators_vals)
plt.xlabel('n_estimators (default is 100)')
plt.ylabel('Accuracy(0-1)')

plt.show()
```

<h4> The effect of max_features on accuracy: </h4>

```python
max_features_vals = []
indices = []
for m_feature in ['auto', 'sqrt']:
    clf_model_max_features = RandomForestClassifier(max_features=m_feature)   
    clf_model_max_features.fit(train_x, train_y)
    predict_y_max_features = clf_model_max_features.predict(test_x)
    max_features_vals.append(accuracy_score(test_y, predict_y_max_features))
    indices.append(m_feature)
plt.plot(indices, max_features_vals)
plt.xlabel('max_features')
plt.ylabel('Accuracy(0-1)')

plt.show()
```

We can see that random forest has higher accuracy in respect to the decision tree method. This is because random forest takes multiple decision trees into account and therefore has lower bias and higher variance. Similarly, decision tree has higher bias and lower variance. The optimal solution is inbetween these two states.

Note that:

$Total Error = Bias ^ 2 + Variance + Irreducible Error$
